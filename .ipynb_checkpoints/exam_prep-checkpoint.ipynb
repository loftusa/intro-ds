## Distributions

### Definitions

```python
# uniform
def uniform(a,b):
    return 1/(b-a)

def gaussian(X, mu=0, var=1):
    sigma = np.sqrt(var)
    k = 1 / np.sqrt(2*pi*var)
    exp = np.exp(-(X-mu)**2 / (2*var))
    return k * exp

x = np.linspace(-5, 5, 1000)
plt.plot(x, gaussian(x));
```

### Summary stats

```python
def E(x, f):
    return np.trapz(x * f(x), x)

def Moment(k, x, f):
    return E(x, f) ** k

def inner(data: np.ndarray):
    mu = np.mean(data)
    var = np.var(data)
    sigma = np.sqrt(var)
    return (data - mu) / sigma

def skewness(data: np.ndarray):
    scaled = inner(data)
    return np.mean(scaled**3)

def kurtosis(data: np.ndarray):
    scaled = inner(data)
    return np.mean(scaled**4)
```

## Least Squares

```python
# generate sample with error
x = 3 * np.random.rand(50) # between 0 and 3
e = 1 * np.random.randn(x.size) # noise
y = 10*np.cos(x+1) + e;  plt.plot(x,y,'.');
```

```python
def poly_fit(x, y, funcs=[], plot=False, ret=False):
    X = np.ones((x.size, len(funcs)+1))
    if funcs:
        for i, f in enumerate(funcs):
            X[:, i+1] = f
    
    # black box
    Xpinv = np.linalg.inv(X.T @ X) @ X.T
    H = X @ Xpinv
    yhat = H @ y
    
    if plot:
        fit = np.vstack((x, yhat))
        fit = fit[:, np.argsort(x)]
        
        plt.plot(x, y, 'k.')
        plt.plot(fit[0], fit[1], 'r')
    
    if ret:
        return yhat
    
# generate sample with error
x = 3 * np.random.rand(50) # between 0 and 3
e = 1 * np.random.randn(x.size) # noise
y = 10*np.cos(x+1) + e

poly_fit(x, y, [x, np.sin(x), np.cos(x)], plot=True)
```

## PCA

### Define functions

```python
from scipy.stats import norm as gaussian
import pandas as pd
from numpy import sin, cos, pi
plt.rcParams['figure.figsize'] = 6, 6

def pl(x, y, xlims=[-10, 10], ylims=[-10,10], **kwargs):
    plt.scatter(x, y, **kwargs)
    plt.xlim(*xlims)
    plt.ylim(*ylims)
    
def rot(theta):
    return np.array([[cos(theta), -sin(theta)],
               [sin(theta),  cos(theta)]])
```

### Find principal eigenvector, rotate, scale by eigenvalue

```python
N = gaussian.rvs(0,1,(2,50))
N[0] /= 2
N[1] *= 3

N = (N.T @ rot(pi/4)).T

eig, eigv, _ = np.linalg.svd(N)
eig_ = -eig[:,0]
theta = np.arccos(eig_[0])
N = (N.T @ rot(theta)).T
N[0, :] /= eigv[0]
# N = (N.T @ rot(theta)).T

pl(N[0], N[1], c='k', s=12);
```

### Rotate / scale some normal data, show shift

```python
# generate many 2D (column) vectors 
np.random.seed(42)
N = gaussian.rvs(0,1,(2,50))
# copy and scale it
S = np.copy(N)
S[0,:] *= 4  # scale axis 0
# rotate it
f = +pi/4    # rotate by 45 degrees
R = np.array([[np.cos(f), -np.sin(f)],
           [np.sin(f),  np.cos(f)]]) 
X = R @ S

# shift it
X += np.array([[1],[3]])

# plot
pl(X[0], X[1], c='k')
avg = np.mean(X, axis=1, keepdims=True)
X -= avg
pl(X[0], X[1], c='r')
```

### Plot transformations on X

```python
# sample randomly from bivariate normal
X = np.random.randn(2, 50)
pl(X[0],X[1])

X[0] *= 3
X[1] /= 2
pl(X[0], X[1])

C = X @ X.T / (X.shape[1] - 1)
print(C)

theta = pi / 4
X = (X.T @ rot(theta)).T
pl(X[0], X[1])

C = X @ X.T / (X.shape[1] - 1)
print(C)
```
