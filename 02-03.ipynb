```python
import math
from math import pi


def f(x, k=2):
    return x**k

class Robot:
    def __init__(self, x=0, y=0, angle=0):
        self.x, self.y, self.angle = x, y, angle
        self.path = [(x,y)]
        
    def move(self, length=1):
        self.x += length * math.cos(self.angle)
        self.y += length * math.sin(self.angle)
        self.path.append((self.x, self.y))
        
r = Robot()

r.move(1)
print(r.x)
print(r.y)
r.move(1)
r.move(1)
```
